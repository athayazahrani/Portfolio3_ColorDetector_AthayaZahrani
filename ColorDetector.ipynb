```python
import cv2

# In Bahasa
# Inisiasi kamera
kamera = cv2.VideoCapture(0)
kamera.set(cv2.CAP_PROP_FRAME_WIDTH, 1280)
kamera.set(cv2.CAP_PROP_FRAME_HEIGHT, 720)

def get_color_name(hue, saturation, value):
    """Determine the color name based on HSV values."""
    if hue == 0 and saturation == 0:
        return "WHITE"  
    elif value < 50:
        return "BLACK"  
    elif saturation < 50:
        return "GRAY"  
    elif hue < 5 or hue >= 170:
        return "RED"  
    elif hue < 15:
        return "ORANGE"  
    elif hue < 35:
        return "YELLOW"  
    elif hue < 50:
        return "YELLOW-GREEN"  
    elif hue < 75:
        return "GREEN"  
    elif hue < 90:
        return "LIGHT GREEN"  
    elif hue < 110:
        return "CYAN"  
    elif hue < 130:
        return "LIGHT BLUE"  
    elif hue < 150:
        return "BLUE"  
    elif hue < 165:
        return "PURPLE"  
    elif hue < 170:
        return "PINK"  
    else:
        return "RED"  

while True:
    # Menangkan frame 
    ret, frame = kamera.read()
    if not ret:
        break

    # Konversi dari BGR ke HSV
    hsv_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    height, width, _ = frame.shape

    # Menentukan titik tengah frame
    cx, cy = width // 2, height // 2

    # Menentukan nilai HSV dari titik tengah
    pixel_center = hsv_frame[cy, cx]
    hue, saturation, value = pixel_center

    # Menentukan nama warna
    color = get_color_name(hue, saturation, value)

    # Menentukan nilai BGR dari titik tengah untuk warna lingkaran dan teks
    pixel_center_bgr = frame[cy, cx]
    b, g, r = map(int, pixel_center_bgr)

    # Menggambar lingkaran di tengah frame dengan warna yang terdeteksi
    cv2.circle(frame, (cx, cy), 20, (b, g, r), -1)  # Gunakan warna yang terdeteksi

    # Menampilkan nama warna pada frame dengan warna yang terdeteksi dan font yang dipilih
    cv2.putText(frame, color, (cx - 100, cy - 100), cv2.FONT_HERSHEY_TRIPLEX, 1.5, (b, g, r), 2)

    # Menampilkan frame dengan judul pendeteksi warna
    cv2.imshow("Color Detector Programme", frame)

    # Keluar dari loop ketika 'ESC' ditekan
    if cv2.waitKey(1) & 0xFF == 27:
        break

# Melepaskan kamera dan menutup semua jendela
kamera.release()
cv2.destroyAllWindows()
```
